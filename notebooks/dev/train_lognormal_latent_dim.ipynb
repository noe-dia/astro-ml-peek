```python
from astro_peek.identifier import CCA_Score 

@torch.no_grad()
def get_cca_score_no_repeat(encoder_x_list, encoder_y_list, x_set = x_set_1, y_set = y_set_1):
    latents_x = []
    latents_y = []
    for (encoder_x, encoder_y) in zip(encoder_x_list, encoder_y_list):
        latents_x_list = []
        latents_y_list = []
        for (x, y) in zip(x_set, y_set):
            latents_x_list.append(encoder_x(x.unsqueeze(0).to(device)).to("cpu").squeeze())     
            latents_y_list.append(encoder_y(y.to(device)).to("cpu")) 

        latents_x.append(np.array(latents_x_list))
        latents_y.append(np.array(latents_y_list))
    latents_x = np.array(latents_x)
    latents_y = np.array(latents_y)
    def get_cca_score(latents): 
        cca_metric = CCA_Score(np.array(latents))
        return cca_metric.compute_all_pairwise_cca()
    
    cca_score_x = get_cca_score(latents_x)
    cca_score_y = get_cca_score(latents_y)
    print("CCA Score feature encoder", cca_score_x)
    print("CCA Score label encoder", cca_score_y)
    latent_dim = latents_x.shape[-1]
    
    from itertools import combinations 
    
    possible_pairs_idx = list(combinations(np.arange(len(latents_x)), r = 2)) # just 2 pairs of models at a time
    cca_score_x_no_repeat = []
    cca_score_y_no_repeat = []
    for pair in possible_pairs_idx: 
        cca_score_x_no_repeat.append(cca_score_x[pair])
        cca_score_y_no_repeat.append(cca_score_y[pair])
    
    
    cca_score_x_no_repeat = np.array(cca_score_x_no_repeat)
    cca_score_y_no_repeat = np.array(cca_score_y_no_repeat)

    return cca_score_x_no_repeat, cca_score_y_no_repeat
    

latent_dims = [3, 5, 32]
encoder_x_full_list = [encoder_x_latent_dim_1, encoder_x_latent_dim_2, encoder_x_latent_dim_32]
encoder_y_full_list = [encoder_y_latent_dim_1, encoder_y_latent_dim_2, encoder_y_latent_dim_32]

for i, latent_dim in enumerate(latent_dims): 
    cca_score_x_no_repeat, cca_score_y_no_repeat = get_cca_score_no_repeat(encoder_x_full_list[i], encoder_y_full_list[i])
    cca_scores_x[latent_dim] = cca_score_x_no_repeat.mean()
    cca_scores_y[latent_dim] = cca_score_y_no_repeat.mean()
    
```

Computing the $R^2$ score 

```python
from torch import nn
from torch import optim
class R2_score: 
    def __init__(self, latent_list): 
        """
        Wrapper class to compute a R2 score in a list of latent vectors.  
        latent_list = list of latent vectors produced by different models. Shape: (nmodels, nsamples, ndims) 
        """
    #TODO
        self.ndims = latent_list.shape[-1]
        self.latent_list = latent_list
        self.device = "cuda" if torch.cuda.is_available() else "cpu"

    def _fit_matrix(self, latent_a, latent_b, num_iter = 10000, lr = 1e-2):
        linear_model = nn.Linear(self.ndims, self.ndims, bias = False).to(self.device)
        optimizer = optim.Adam(linear_model.parameters(), lr = lr)
        loss_fn = nn.MSELoss(reduction = "mean")
        
        for i in tqdm(range(num_iter)): 
            optimizer.zero_grad()
            predicted_latent_b = linear_model(torch.tensor(latent_a).to(self.device))
            loss = loss_fn(torch.tensor(latent_b).to(self.device), predicted_latent_b)
            loss.backward()
            optimizer.step()
        print(f"Final loss, fit matrix: {loss}")
        return linear_model, loss

    def compute_pair_score(self, latent_a, latent_b, fitter_args = {}): 
        """
        latent_a and latent_b are (nsamples, ndims)
        Computes the R2 score for a pair of self.latent_list produced by two models M_1 and M_2
        """
        
        linear_model, loss = self._fit_matrix(latent_a, latent_b, **fitter_args)
        predicted_latent_b = linear_model(torch.tensor(latent_a).to(self.device)).detach().cpu().numpy()
        ss_res = np.sum((latent_b - predicted_latent_b) ** 2)
        ss_tot = np.sum((latent_b - np.mean(latent_b, axis = 0)) ** 2)
        print(ss_res)
        print(ss_tot)
        r2 = 1 - ss_res/ss_tot
        print(r2)
        return r2, linear_model
        
    def compute_score(self, latent_a, latent_b):
        """
        latent_a: (n, d)
        latent_b: (n, d)
        """
    
        # Solve Y ≈ X A
        A = np.linalg.lstsq(latent_a, latent_b, rcond=None)[0]
    
        predicted_latent_b = latent_a @ A
    
        ss_res = np.sum((latent_b - predicted_latent_b) ** 2)
        ss_tot = np.sum((latent_b - np.mean(latent_b, axis=0)) ** 2)
    
        r2 = 1 - ss_res / ss_tot
    
        return r2, A, None
    
```
