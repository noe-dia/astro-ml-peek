```python
from datasets import load_from_disk
from tqdm import tqdm
from astro_peek.datasets.make_rings import make_rings
from astro_peek.datasets.make_lenses import simulate_system

np.random.seed(42)
dataset_size = 1000

prior_radius_a = lambda size: np.random.normal(loc = 0.9, scale = 0.01, size = size)
prior_radius_b = lambda size: np.random.normal(loc = 1.2, scale = 0.01, size = size)

prior_phi_a = lambda size: np.random.normal(loc = -np.pi/3, scale = 0.05, size = size)
prior_phi_b = lambda size: np.random.normal(loc = 0, scale = 0.05, size = size)

# Creating latent factors and normalizing them...
radius = np.concatenate([sample_prior(prior_radius_a, dataset_size), sample_prior(prior_radius_b, dataset_size)])
rng = np.random.default_rng(seed=42)
rng.shuffle(radius)

sigma = np.concatenate([sample_prior(prior_phi_a, dataset_size), sample_prior(prior_phi_b, dataset_size)])
rng = np.random.default_rng(seed=42)
rng.shuffle(sigma)

labels = np.concatenate([np.zeros(dataset_size), np.ones(dataset_size)])
rng = np.random.default_rng(seed=42)
rng.shuffle(labels)

theta = np.stack([radius, sigma]).T

# Creating simulations from it 
images = []
for params in tqdm(theta): 
    image, _ = simulate_system(*params)
    images.append(image) 

images = np.array(images)
# radius = (radius - radius_prior["low"]) / (radius_prior["high"] - radius_prior["low"])
# sigma =  (sigma - sigma_prior["low"]) / (sigma_prior["high"] - sigma_prior["low"])
theta = np.stack([radius, sigma]).T
```

Computing the $R^2$ score 

```python
from torch import nn
from torch import optim
class R2_score: 
    def __init__(self, latent_list): 
        """
        Wrapper class to compute a R2 score in a list of latent vectors.  
        latent_list = list of latent vectors produced by different models. Shape: (nmodels, nsamples, ndims) 
        """
    #TODO
        self.ndims = latent_list.shape[-1]
        self.latent_list = latent_list
        self.device = "cuda" if torch.cuda.is_available() else "cpu"

    def _fit_matrix(self, latent_a, latent_b, num_iter = 1000, lr = 1e-1):
        linear_model = nn.Linear(self.ndims, self.ndims).to(self.device)
        optimizer = optim.Adam(linear_model.parameters(), lr = lr)
        loss_fn = nn.MSELoss(reduction = "mean")
        
        for i in tqdm(range(num_iter)): 
            optimizer.zero_grad()
            predicted_latent_b = linear_model(torch.tensor(latent_a).to(self.device))
            loss = loss_fn(torch.tensor(latent_b).to(self.device), predicted_latent_b)
            loss.backward()
            optimizer.step()
        print(f"Final loss, fit matrix: {loss}")
        return linear_model, loss

    def compute_pair_score(self, latent_a, latent_b, fitter_args = {}): 
        """
        latent_a and latent_b are (nsamples, ndims)
        Computes the R2 score for a pair of self.latent_list produced by two models M_1 and M_2
        """
        
        linear_model, loss = self._fit_matrix(latent_a, latent_b, **fitter_args)
        predicted_latent_b = linear_model(torch.tensor(latent_a).to(self.device)).detach().cpu().numpy()
        ss_res = np.sum((latent_b - predicted_latent_b) ** 2)
        ss_tot = np.sum((latent_b - np.mean(latent_b, axis = 0)) ** 2)
        r2 = 1 - ss_res/ss_tot
        print(r2.shape)
        return r2, linear_model
        
    def compute_score(self, fitter_args = {}): 
        """
        Computes the R2 scores between every possible pair of self.latent_list 

        Output: (len(latent_list), ) = R2 scores.
        """
        # Computing every possible pairs 
        possible_pairs_idx = list(combinations(np.arange(len(self.latent_list)), r = 2)) # just 2 pairs of models at a time
        scores = np.empty(shape = (len(self.latent_list),), dtype = np.float32)
        linear_model_list = []
        
        for i, possible_pair in enumerate(possible_pairs_idx): 
            latent_1 = self.latent_list[possible_pair[0]]
            latent_2 = self.latent_list[possible_pair[1]]
            scores[i], linear_model = self.compute_pair_score(latent_1, latent_2, **fitter_args)
            linear_model_list.append(linear_model)

        return scores, linear_model_list, possible_pairs_idx
```
