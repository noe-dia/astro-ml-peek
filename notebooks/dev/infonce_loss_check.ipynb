## Objective: Understanding the model from the exponential family loss function and coding it

$$
p(y \mid x, S) = \frac{\exp(z_x^{T}z_y)}{\sum_{y'\in S} \exp(z_x^{T}z_{y'})}
$$

The loss we want to compute is obtained by computing the negative log-likelihood $\log p(y\mid x, S)$. 
$$
\begin{align}
\log p(y\mid x, S) &= z_x^{T}z_y - \log \left(\sum_{y'\in S} \exp (z_x^{T} z_y')\right) \quad [\text{for one pair (x, y)}]
\\
\mathcal{L}&= - \frac{1}{B} \sum_{i=1}^{B} \log p(y_i\mid x_i, S)
\end{align}
$$

In principle, S should contain all the possible values that $y$ can take. In continuous settings, this means computing an integral at every step of training which can be expensive. We will further approximate the loss objective by making $S$ stochastic, i.e. by making it the set of possible labels from the batch.  

```python
import itertools
import numpy as np

# Method A: Using Itertools (Best for small sets)
models = np.array([1, 2, 3, 4, 5])
unique_pairs = list(itertools.combinations(models, 2))
```
