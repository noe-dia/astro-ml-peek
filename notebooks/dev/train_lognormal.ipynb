```python
from datasets import load_from_disk 

dset = load_from_disk("/home/noedia/links/scratch/crl/experiments/lognormal/data/lognormal-flat")
dset = dset.with_format("torch")
nsamples = 1_000

x_set_1 = dset["train"][:nsamples]["image"]
x_set_2 = dset["train"][nsamples:2 * nsamples]["image"]

y_set_1 = dset["train"][:nsamples]["theta"]
y_set_2 = dset["train"][nsamples:2 * nsamples]["theta"]
```

```python
from astro_peek.identifier import CCA_Score
import torch
import numpy as np 
latents_x = []
latents_y = []
device = "cuda" if torch.cuda.is_available() else "cpu"
with torch.no_grad():
    for (encoder_x, encoder_y) in zip(encoder_x_list, encoder_y_list):
        latents_x_list = []
        latents_y_list = []
        for (x, y) in zip(x_set_1, y_set_1):
            latents_x_list.append(encoder_x(x.unsqueeze(0).to(device)).to("cpu").squeeze())     
            latents_y_list.append(encoder_y(y.to(device)).to("cpu")) 

        latents_x.append(np.array(latents_x_list))
        latents_y.append(np.array(latents_y_list))
```

```python
latents_x = np.array(latents_x)
latents_y = np.array(latents_y)

def get_cca_score(latents): 
    cca_metric = CCA_Score(np.array(latents))
    return cca_metric.compute_all_pairwise_cca()

cca_score_x = get_cca_score(latents_x)
cca_score_y = get_cca_score(latents_y)
print("CCA Score feature encoder", cca_score_x)
print("CCA Score label encoder", cca_score_y)
```

```python
from itertools import combinations 

possible_pairs_idx = list(combinations(np.arange(len(latents_x)), r = 2)) # just 2 pairs of models at a time
cca_score_x_no_repeat = []
cca_score_y_no_repeat = []
for pair in possible_pairs_idx: 
    cca_score_x_no_repeat.append(cca_score_x[pair])
    cca_score_y_no_repeat.append(cca_score_y[pair])


cca_score_x_no_repeat = np.array(cca_score_x_no_repeat)
cca_score_y_no_repeat = np.array(cca_score_y_no_repeat)
```

```python
import scienceplots 
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
plt.style.use("science")
plt.rcParams.update({
         "font.size": 20,
         'xtick.major.size': 8, 
         'xtick.minor.size': 3,
         'xtick.major.width': 1.5,
         'xtick.minor.width': 1,
         'ytick.major.size': 8,
         'ytick.minor.size': 3,
         'ytick.major.width': 1.5,
         'ytick.minor.width': 1, 
         "lines.linewidth": 2,
        "legend.fontsize": 20,
         "image.cmap": "magma"
        })

fig, axs = plt.subplots(1, 1, figsize = (6, 4))

dims = np.arange(1, latents_x.shape[-1]+1)
colors = ["cornflowerblue", "tomato", "mediumseagreen"]
axs.axhline(cca_score_x_no_repeat.mean(), ls = "--", color = "k", label = "Mean CCA (Feature models)")
axs.axhline(cca_score_y_no_repeat.mean(), ls = "-.", color = "k", label = "Mean CCA (Label models)")

for i, pair in enumerate(possible_pairs_idx[::-1]):
    axs.scatter(dims, cca_score_x[pair], color = colors[i], s = 100)
    axs.scatter(dims, cca_score_y[pair], color = colors[i], marker = "*", s = 100)
    axs.scatter([], [], label = f"Models {list(np.array(pair) + 1)}", color = colors[i])

axs.scatter([], [], label = f"Feature Models", color = "k")
axs.scatter([], [], label = f"Label Models", marker = "*", color = "k")



axs.set(xticks = dims,  ylim = [0., 1.05], xlabel = "Latent Dimension", ylabel = "CCA Score")
axs.legend(fontsize = 13)
plt.savefig("cca_plot_lognormal.pdf", bbox_inches = "tight")
```
